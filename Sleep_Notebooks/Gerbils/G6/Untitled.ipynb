```python
animalPaths = [os.path.join(experimentFolder,item) for  item in os.listdir(experimentFolder) if not os.path.isfile(os.path.join(experimentFolder,item))]
```

```python
for animal in sorted(animalPaths):
    print 'processing %s' %animal
    ncsPaths = []
    for ncsFile in tools.locate('*.ncs',animal):
        ncsPaths.append(os.path.join(ncsFile[0],ncsFile[1]))
    ncsPaths = sorted(ncsPaths)
    ncsPaths = [ path for path in ncsPaths if (path.find('sleep')>0 or path.find('Sleep')>0) ]
    for cscFile in ncsPaths:
        if not os.path.isfile(cscFile.split('.')[0]+'.lfp'):
            lfp = signale.load_ncsFile(cscFile)
            lfp.whitenARMA()
            fileName = cscFile.split('.')[0]+'.lfp'
            pkl.dump(lfp,open(fileName,'wb'),pkl.HIGHEST_PROTOCOL)
            print 'lfp saved to %s' %fileName
```
