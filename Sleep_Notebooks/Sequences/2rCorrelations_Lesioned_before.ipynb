```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as pl
import pickle as pkl
import Recordings
import signale
import extras.trees
import os
```

###### Finding spiketrains and ephys files!¶

```python
ephysList = []
spkList = []
for item in extras.trees.locate('*.spk',lesionAnimalBefore):
    spkList.append(os.path.join(item[0],item[1]))
for item in extras.trees.locate('*.ephys',lesionAnimalBefore):
    ephysList.append(os.path.join(item[0],item[1]))
#####################################################
for ii in range(len(spkList)) :
    spk = pkl.load(open(spkList[ii],'rb'))
    rec = pkl.load(open(ephysList[ii],'rb'))
    before = Recordings.ephys(spk,rec.LFPs)
    before.spikes.timeAxis = before.timeAxis.copy()
    del rec
```

```python
interID = np.array([ 7, 21, 23, 27, 57, 68, 80, 88]) #This came from analysis of after session!
```

```python
lfpB = before.LFPs[9]
lfpB.ripple_recorder()
```

```python
before.LFPs.ripples = lfpB.ripples
before.rippleSpikeContent(method='fixed_window')
```

```python
#Removing Interneurons from spikeList

for item in before.ripple_MUA:
    for jj,jtem in enumerate(item):
        if jj in interID:
            del item[jj]
print len(before.ripple_MUA[0])
```

```python
rippStartB = np.array([item[0] for item in before.LFPs.ripples])
rippEndB =   np.array([item[1] for item in before.LFPs.ripples])
rippPeakB =  np.array([item[3] for item in before.LFPs.ripples])
```

###### Fixed ripple window

```python
N_rows = len(before.ripple_MUA[0])
binary_ST = []
for ii,zug in enumerate(before.ripple_MUA):
    binary_p_zug = np.zeros([N_rows,100])
    for jj in range(N_rows):
        if zug[jj].size > 0:
            for item in zug[jj]:
                idx = np.int64(item - np.int64(rippPeakB[ii]) + 50)
                binary_p_zug[jj,idx] += 1
               
    binary_ST.append(binary_p_zug)
```

```python
rateStack = []
for item in binary_ST:
    rate = []
    for ii in range(98):
        rate.append(np.array([item[ii][:50].sum(),item[ii][50:].sum()]))
    rateStack.append(np.array(rate))
```

```python
rateStack[0][0]
```

```python
corrMatStack = []
for item in rateStack:
    corrMat = np.zeros([N_rows,N_rows])
    for ii in range(N_rows):
        for jj in range(ii,N_rows):
            corcoeff = 0.0
            corcoeff = np.corrcoef(item[ii,:],item[jj,:])[0,1]
            corrMat[ii,jj] = corcoeff
            corrMat[jj,ii] = corcoeff
    corrMatStack.append(corrMat)
```

```python
for item in rateStack:
    print np.corrcoef(item[10,:],item[11,:])[0,1]
```

```python
nanMean = []
nanSD = []
nanSUM = []
for item in corrMatStack:
    nanMean.append(np.nanmean(item))
    nanSD.append(np.nanstd(item))
    nanSUM.append(np.nansum(item))
```

```python
pl.hist(nanSUM,40);
```

```python
lowCorr = np.intersect1d(np.where(np.array(nanSUM) > 0.01)[0],np.where(np.array(nanSUM) < 50)[0])
highCorr = np.where(np.array(nanSUM) > 50)[0]
```
