```python
ephysList = []
spkList = []
for item in extras.trees.locate('*.spk',lesionAnimalAfter):
    spkList.append(os.path.join(item[0],item[1]))
for item in extras.trees.locate('*.ephys',lesionAnimalAfter):
    ephysList.append(os.path.join(item[0],item[1]))
```

```python
for ii in range(len(spkList)) :
    spk = pkl.load(open(spkList[ii],'rb'))
    rec = pkl.load(open(ephysList[ii],'rb'))
    recording = Recordings.ephys(spk,rec.LFPs)
    recording.spikes.timeAxis = recording.timeAxis.copy()
    del rec
```

```python
recording.LFPs.ripples = lfp.ripples
recording.rippleSpikeContent(method='fixed_window')
```

```python
meanRates = np.array(recording.spikes.mean_rates())
interID = np.where(meanRates > 4)[0]
print interID
print meanRates[interID]
#Removing Interneurons from spikeList

for item in recording.ripple_MUA:
    for jj,jtem in enumerate(item):
        if jj in interID:
            del item[jj]
print len(recording.ripple_MUA[0])
```

```python
rippStart = np.array([item[0] for item in recording.LFPs.ripples])
rippEnd =   np.array([item[1] for item in recording.LFPs.ripples])
rippPeak =   np.array([item[3] for item in recording.LFPs.ripples])
```

###### Dynamic ripple window

```python
rippDuration = rippEnd - rippStart
N_rows = recording.spikes.id_list.size
binary_ST = []
for ii in range(rippDuration.size):
    binary_p_zug = np.zeros([N_rows,np.int16(np.round(rippDuration[ii])) + 1])
    zug = recording.ripple_MUA[ii]
    for jj in range(N_rows):
        if zug[jj].size > 0:
            for item in zug[jj]:
                idx = np.int16(item - rippStart[ii])
                binary_p_zug[jj,idx] += 1
               
    binary_ST.append(binary_p_zug)
```

###### Fixed ripple window

```python
N_rows = len(recording.ripple_MUA[0])
binary_ST = []
for ii in range(rippPeak.size):
    binary_p_zug = np.zeros([N_rows,100])
    zug = recording.ripple_MUA[ii]
    for jj in range(N_rows):
        if zug[jj].size > 0:
            for item in zug[jj]:
                idx = np.int16(item - np.int64(rippPeak[ii]) + 50)
                binary_p_zug[jj,idx] += 1
               
    binary_ST.append(binary_p_zug)
```

```python
corrMatStack = []
for item in binary_ST:
    corrMat = np.zeros([N_rows,N_rows])
    for ii in range(N_rows):
        for jj in range(ii,N_rows):
            corcoeff = 0.0
            corcoeff = np.corrcoef(item[ii,:],item[jj,:])[0,1]
            corrMat[ii,jj] = corcoeff
            corrMat[jj,ii] = corcoeff
    corrMatStack.append(corrMat)
```

```python
nanMean = []
nanSD = []
nanSUM = []
for item in corrMatStack:
    nanMean.append(np.nanmean(item))
    nanSD.append(np.nanstd(item))
    nanSUM.append(np.nansum(item))
```

```python
lowCorr = np.intersect1d(np.where(np.array(nanSUM) > 5)[0],np.where(np.array(nanSUM) < 30)[0])
highCorr = np.where(np.array(nanSUM) > 30)[0]
```

```python
HcorrVec = []
LcorrVec = []
for item in highCorr:
    vec = []
    for ii in range(106):
        for jj in range(ii+1,98):
            vec.append(corrMatStack[item][ii,jj])
    HcorrVec.append(np.array(vec))

for item in lowCorr:
    vec = []
    for ii in range(106):
        for jj in range(ii+1,98):
            vec.append(corrMatStack[item][ii,jj])
    LcorrVec.append(np.array(vec))

########################################
HNN = len(HcorrVec)
LNN = len(LcorrVec)
HhyperAngle = []
LhyperAngle = []
for ii in range(HNN):
    for jj in range(ii+1,HNN):
        inan = np.nan_to_num(HcorrVec[ii])
        jnan = np.nan_to_num(HcorrVec[jj])
        idotj = np.dot(inan,jnan)
        il = np.math.sqrt(np.dot(inan,inan))
        jl = np.math.sqrt(np.dot(jnan,jnan))
        HhyperAngle.append(np.math.acos(idotj/il/jl)*(180/np.math.pi))
        
for ii in range(LNN):
    for jj in range(ii+1,LNN):
        inan = np.nan_to_num(LcorrVec[ii])
        jnan = np.nan_to_num(LcorrVec[jj])
        idotj = np.dot(inan,jnan)
        il = np.math.sqrt(np.dot(inan,inan))
        jl = np.math.sqrt(np.dot(jnan,jnan))
        LhyperAngle.append(np.math.acos(idotj/il/jl)*(180/np.math.pi))
```

```python
f,ax = pl.subplots(1,1,figsize = [10,5])
ax.hist([HhyperAngle,LhyperAngle],100,normed=False,log=False);
pl.legend();
pl.xlim([60,110])
```

```python
HNN = len(HcorrVec)
LNN = len(LcorrVec)
HhyperDistance = []
LhyperDistance = []
for ii in range(HNN):
    for jj in range(ii+1,HNN):
        iminusj = np.subtract(HcorrVec[ii],HcorrVec[jj])
        iminusjNan = np.nan_to_num(iminusj)
        #hyperDistance.append(np.math.sqrt(np.dot(iminusjNan,iminusjNan)))
        HhyperDistance.append(np.linalg.norm(iminusjNan))

for ii in range(LNN):
    for jj in range(ii+1,LNN):
        iminusj = np.subtract(LcorrVec[ii],LcorrVec[jj])
        iminusjNan = np.nan_to_num(iminusj)
        #hyperDistance.append(np.math.sqrt(np.dot(iminusjNan,iminusjNan)))
        LhyperDistance.append(np.linalg.norm(iminusjNan))
```
