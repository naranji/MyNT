```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as pl
import pickle as pkl
import Recordings
import signale
import extras.trees
import os
```

```python
WorkPath = '/mnt/Data/IpythonNotebooks/PCA_MEC/'
Pre =   '/mnt/Data/ephysdata/Rats/MECLesion/Rat434Lesion/sleep5/'
Track = '/mnt/Data/ephysdata/Rats/MECLesion/Rat434Lesion/lineartrack1/'
Post =  '/mnt/Data/ephysdata/Rats/MECLesion/Rat434Lesion/sleep6/'
```

```python
ExperimentSessions = [Pre,Track,Post]
spkList = []
ephysList = []
pllSpkTrns = []
for item in ExperimentSessions:
    for jtem in extras.trees.locate('*.spk',item):
        spkList.append(os.path.join(jtem[0],jtem[1]))
    for jtem in extras.trees.locate('*.ephys',item):
        ephysList.append(os.path.join(jtem[0],jtem[1]))
#####################################################
for ii in range(len(spkList)) :
    spk = pkl.load(open(spkList[ii],'rb'))
    rec = pkl.load(open(ephysList[ii],'rb'))
    before = Recordings.ephys(spk,rec.LFPs)
    spk.timeAxis = rec.timeAxis.copy()
    pllSpkTrns.append(spk)
    del spk
    del rec
```

```python
corrMatrices = []
Qs = []
eigVals = []
eigVecs = []
Ps = []
for item in pllSpkTrns:
    Q,C,l,V,P = item.PCA()
    corrMatrices.append(C)
    Qs.append(Q)
    eigVals.append(l)
    eigVecs.append(V)
    Ps.append(P)
```
