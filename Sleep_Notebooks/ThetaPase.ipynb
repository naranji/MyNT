```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as pl
import pickle as pkl
import Recordings
import signale
import extras.trees
import os
```

```python
lesionAnimalLT = '/media/chenani/Data/ephysdata/Rats/MECLesion/Rat434Lesion/lineartrack1/'
```

```python
ephysList = []
spkList = []
for item in extras.trees.locate('*1.spk',lesionAnimalLT):
    spkList.append(os.path.join(item[0],item[1]))
for item in extras.trees.locate('*1.ephys',lesionAnimalLT):
    ephysList.append(os.path.join(item[0],item[1]))
#####################################################
for ii in range(len(spkList)) :
    spk = pkl.load(open(spkList[ii],'rb'))
    rec = pkl.load(open(ephysList[ii],'rb'))
    linearTrack = Recordings.ephys(spk,rec.LFPs)
    linearTrack.spikes.timeAxis = linearTrack.timeAxis.copy()
    del rec
```

```python
f,(ax1,ax2) = pl.subplots(1,2,figsize= [20,30])
linearTrack.LFPs.plot(f,ax1)
linearTrack.LFPs.fft_plot(fig=f,ax=ax2)
ax2.set_xlim([0,20])
```
